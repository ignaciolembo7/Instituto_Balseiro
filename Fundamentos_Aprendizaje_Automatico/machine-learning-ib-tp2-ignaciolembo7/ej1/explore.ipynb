# Explore Data

Importamos los paquetes necesarios y los Datasets que exploraremos.

```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Función para cargar los datos desde el archivo pickle
def load_data(file_path):
    with open(file_path, 'rb') as f:
        data = pickle.load(f)
    return data
dataset_1 = load_data("dataset_1.pkl")
dataset_2 = load_data("dataset_2.pkl")
dataset_3 = load_data("dataset_3.pkl")
```

Visualizar los datos cargados

# Dataset 1

```python
print("Dataset 1:")
print(dataset_1.keys())  # Ver las claves del diccionario
print("Número de características en dataset 1:", dataset_1["X"].shape[1])
print("Número de ejemplos en dataset 1:", dataset_1["X"].shape[0])
```

```python
# Extraer características (X) y variable objetivo (y) del dataset
X = dataset_1["X"]
y = dataset_1["y"]

# Plotear X vs y
num_rows = 2
num_cols = 5

fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 10))

axes = axes.flatten()
for i in range(X.shape[1]):
    ax = axes[i]
    X_column = X[:, i]
    ax.scatter(X_column, y, color='blue', alpha=0.5)
    ax.set_title(f'Columna {i+1}')
    ax.set_xlabel('Características (X)')
    ax.set_ylabel('Variable Objetivo (y)')
    ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
X_df = pd.DataFrame(dataset_1["X"], columns=[f'Característica {i+1}' for i in range(X.shape[1])])
X_df.describe()
```

```python
correlation_matrix = X_df.corr()
numerical_columns = X_df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numerical_columns.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".4f"), 
plt.show()
```

Este tipo de datos presenta un comportamiento continuo aproximadamente lineal frente a las variables de entrada, por lo tanto, es convieniente utilizar el modelo LinearRegressor.
