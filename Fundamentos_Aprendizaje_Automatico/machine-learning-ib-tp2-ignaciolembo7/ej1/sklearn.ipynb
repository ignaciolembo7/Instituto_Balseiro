# SKLEARN

Cargo los Datasets e importo los paquetes que vamos a utilizar.

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble  import GradientBoostingRegressor 
from sklearn.metrics import mean_squared_error, accuracy_score

# Función para cargar los datos desde el archivo pickle
def load_data(file_path):
    with open(file_path, 'rb') as f:
        data = pickle.load(f)
    return data
dataset_1 = load_data("dataset_1.pkl")
dataset_2 = load_data("dataset_2.pkl")
dataset_3 = load_data("dataset_3.pkl")
```

Preparo los conjuntos de entrenamiento y validación para cada Dataset.

```python
X_1, y_1 = dataset_1["X"], dataset_1["y"]
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_1, y_1, test_size=0.2, random_state=42)

X_2, y_2 = dataset_2["X"], dataset_2["y"]
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_2, y_2, test_size=0.2, random_state=42)

X_3, y_3 = dataset_3["X"], dataset_3["y"]
X_train_3, X_test_3, y_train_3, y_test_3 = train_test_split(X_3, y_3, test_size=0.2, random_state=42)
```

Entreno y evaluo el modelo LinearRegression para Dataset_1 y Dataset_3.

```python
linear_model_1 = LinearRegression()
linear_model_1.fit(X_train_1, y_train_1)
linear_predictions_1 = linear_model_1.predict(X_test_1)
linear_mse_1 = mean_squared_error(y_test_1, linear_predictions_1)
print("Dataset 1 - LinearRegression test MSE:", linear_mse_1)
num_rows = 2
num_cols = 5
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 10))
axes = axes.flatten()
for i in range(X_test_1.shape[1]):
    ax = axes[i]
    ax.scatter(X_test_1[:, i], y_test_1, color='black', alpha = 0.5, label = 'Conjunto de validación')
    ax.scatter(X_test_1[:, i], linear_predictions_1, color='red', alpha = 0.25, label = 'Predicción del conjunto de validación')
    ax.legend(loc = 'best')
    ax.set_xlabel(f'Feature {i+1}  [X]')
    ax.set_ylabel(f'Target [y]')
    ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
linear_model_3 = LinearRegression()
linear_model_3.fit(X_train_3, y_train_3)
linear_predictions_3 = linear_model_3.predict(X_test_3)
linear_mse_3 = mean_squared_error(y_test_3, linear_predictions_3)
print("Dataset 3 - LinearRegression test MSE:", linear_mse_3)

num_rows = 2  # Redondeo hacia arriba de num_columns/2
num_cols = 5
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 10))
axes = axes.flatten()
for i in range(X_test_3.shape[1]):
    ax = axes[i]
    ax.scatter(X_test_3[:, i], y_test_3, color='black', alpha = 0.5, label = 'Conjunto de testeo')
    ax.scatter(X_test_3[:, i], linear_predictions_3, color='red', alpha = 0.25, label = 'Predicción')
    ax.legend(loc = 'best')
    ax.set_xlabel(f'Feature {i+1}  [X]')
    ax.set_ylabel(f'Target [y]')
    ax.grid(True)
plt.tight_layout()
plt.show()
```

Entreno y evaluo el modelo de LogisticRegression para el Dataset_2.

```python
logistic_model_2 = LogisticRegression()
logistic_model_2.fit(X_train_2, y_train_2)
logistic_predictions_2 = logistic_model_2.predict(X_test_2)
logistic_accuracy_2 = accuracy_score(y_test_2, logistic_predictions_2)
print("Dataset 2 - LogisticClassifier test accuracy:", logistic_accuracy_2)

num_rows = 2
num_cols = 5
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 10))
axes = axes.flatten()
for i in range(X_test_2.shape[1]):
    ax = axes[i]
    ax.scatter(X_test_2[:, i], y_test_2, color='black', alpha = 0.5, label = 'Conjunto de testeo')
    ax.scatter(X_test_2[:, i], logistic_predictions_2, color='red', alpha = 0.25, label = 'Predicción')
    ax.legend(loc = 'best')
    ax.set_xlabel(f'Feature {i+1}  [X]')
    ax.set_ylabel(f'Target [y]')
    ax.grid(True)
plt.tight_layout()
plt.show()
```

Entreno y evaluo el modelo de GradientBoostingRegressor para el Dataset_3.

```python
GBR_3 = GradientBoostingRegressor(learning_rate=1)
GBR_3.fit(X_train_3, y_train_3)
GBR_predictions_3 = GBR_3.predict(X_test_3)
GBR_mse_3 = mean_squared_error(y_test_3, GBR_predictions_3)
print("Dataset 3 - LinearRegression test MSE:", GBR_mse_3)
num_rows = 2 
num_cols = 5
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 10))
axes = axes.flatten()
for i in range(X_test_3.shape[1]):
    ax = axes[i]
    ax.scatter(X_test_3[:, i], y_test_3, color='black', alpha = 0.5, label = 'Conjunto de testeo')
    ax.scatter(X_test_3[:, i], GBR_predictions_3, color='red', alpha = 0.25, label = 'Predicción')
    ax.legend(loc = 'best')
    ax.set_xlabel(f'Feature {i+1}  [X]')
    ax.set_ylabel(f'Target [y]')
    ax.grid(True)
plt.tight_layout()
plt.show()
```

- En primer lugar, podemos ver que los valores de MSE obtenidos utilizando el modelo LinearRegressor de sklearn en los Datasets 1 y 3 son extremadamente similares a los implementados por nosotros en train.ipynb. Lo mismo ocurre con el accuracy obtenido para el Dataset_2 con el modelo LogisticClassifier implementado por nosotros y el de sklearn.

- Además, utilizando los mismos parámetros para comparar los GradientBoostingRegressor (n_estimators=100, learning_rate=1), obtuvimos que los MSE obtenidos son similares, no obstante el tiempo de ejecución de nuestra implementación duró aproximadamente 9 minutos mientras que utilizando scikit-learn tomó 5 segundos. 

- Las diferencias entre los métodos implementados por nosotros y los obtenidos de la librería scikit-learn, en primer lugar, consisten en que los modelos en scikit-learn están implementados de manera eficiente y optimizada, utilizando técnicas avanzadas de programación para garantizar un rendimiento óptimo. Nuestra implementación es más básica o menos optimizada en comparación con las implementaciones de scikit-learn, lo que afecta el rendimiento y la velocidad de los modelos. Además, algunos modelos en scikit-learn pueden tener regularización incorporada por defecto y pueden realizar una mejor optimización de hiperparámetros, lo que también podría llevar a mejoras en el rendimiento. En nuestro caso, la mayor diferencia se nota en la velocidad con la que se ajustan los datos.
