```python
from egtplot import plot_static

C = 2
G = 1

payoff_entries = [
    [(G-C)*0.5], [G], [G],
    [0], [G*0.5], [0],
    [0], [G], [G*0.5]
]

labels = ['Halcones', 'Palomas', 'Bravucones']
simplex = plot_static(payoff_entries, vert_labels=labels, display_parameters=False, paths = False, background= False)
```
